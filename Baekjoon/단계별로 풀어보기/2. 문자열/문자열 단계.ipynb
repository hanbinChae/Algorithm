문자열 개수 입력 : n
반복할 횟수 : num
해당 문자열 : word

```python
n = int(input())
num = []
word = []

for i in range(0,n):
    a,b = map(str,input().split())
    num.append(int(a))
    word.append(b)

for n in range(n):
    for j in range(0,len(word[n])):
        print(num[n]*word[n][j],end="")
    print()
```

가장 많이 쓰인 알파벳 대문자로 출력

```python
word = input().lower()
word_set = list(set(word))
result = []

for i in word_set:
    result.append((word.count(i)))

if result.count(max(result)) >=2:
    print('?')
else:
    print(word_set[result.index(max(result))].upper())
```

숫자 뒤집어서 비교하기

```python
a,b = list(map(str,input().split()))
a = a[::-1]
b = b[::-1]

if int(a)>int(b):
    print(a)
else:
    print(b)
```

다이얼

```python
dial = ['ABC','DEF','GHI','JKL','MNO','PQRS','TUV','WXYZ']
a = list(input())
result = 0

for i in a:
    for j in dial:
        if i in j:
            result+=dial.index(j)+3

print(result)
```

크로아티아 알파벳

```python
n = int(input())
cnt = n

for i in range(n):
    word = input()
    for j in range(0,len(word)-1):
        if word[j] == word[j+1]:
            pass
        elif word[j] in word[j+1:]:
            cnt -= 1
            break;
print(cnt)
```
